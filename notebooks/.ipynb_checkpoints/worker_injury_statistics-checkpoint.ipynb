## Mean number of injuries for a company

We model the mean number of injuries for a given company as an extension of the binomial distribution, the poisson binomial distribution, where we have $n$ independent trials (employees), but not identitically distributed ($p(\text{non-fatal injury})$ for each employee is different). 

Let $p_i$ be the probability that the $i$th employee of the company experiences a non-fatal injury. The mean number of non-fatal injuries in the company are as follows:

\begin{align}
    E[\text{injury} \vert \text{company}]
    &= \sum^{n}_{i = 1}p_i \\
\end{align}

```python
def expected_number_of_injuries(
    employees: List[Employee],
    rates_age_by_occupation: pd.DataFrame = rates_age_by_occupation,
) -> Tuple[float]:
    """Calculates the expected number of injuries
    in an organization given a list of employee objects
    and their respective ages

    Args:
        employees: (List[Employee]) List of employee objects
        rates_age_by_occupation: CPD table giving the p(injury, age_group | occupation)

    Returns:
        Tuple[float]:
            Lower bound: One standard deviation below mean
            Mean: Expected number of employees injured in a year
            Upper bound: One standard deviation above mean
    """
    # Key: (age bucket as int), Value: (age bucket as string)
    int_to_bin_map = dict(enumerate(rates_age_by_occupation.columns[1:].tolist()))

    # Which (age bucket as int) each employee belongs to
    age_group_of_employees = [
        int_to_bin_map[age_bucket_as_int - 1]
        for age_bucket_as_int in np.digitize(
            x=list(map(lambda x: x.age, employees)),
            bins=[14, 16, 20, 25, 35, 45, 55, 65, 100],
        )
    ]

    # Probability of non-fatal injury for each employee
    probability_of_non_fatal_injury_for_all_employees = [
        rates_age_by_occupation.loc[occ_code, age_group]
        for occ_code, age_group in zip(
            list(map(lambda x: x.occ_code, employees)), age_group_of_employees
        )
    ]

    μ = sum(probability_of_non_fatal_injury_for_all_employees)
    σ = np.array(probability_of_non_fatal_injury_for_all_employees) @ (
        1 - np.array(probability_of_non_fatal_injury_for_all_employees)
    )

    # Expected number of injuries in company
    return max(0, μ - σ), μ, μ + σ
```

```python
def expected_number_of_injuries_per_category_for_occupation(
    employees: List[Employee],
    rates_category_by_occupation: pd.DataFrame = rates_body_part_by_occupation,
) -> pandas.DataFrame:
    """Takes a list of employees and calculates the average number of
    injuries per body part in a year using the number of employees in
    each occupation

    Args:
        employees: (List[Employee]) List of employee objects
        rates_category_by_occupation: CPD table giving the p(injury, body_part | occupation),
            or p(injury, nature_of_injury | occupation),
            or p(injury, number_of_days_away_from_work | occupation)

    Returns:
        pandas.DataFrame:
            Table with occupations in the rows and the body part injury type in column,
            cells being expected number of injuries in each category
    """
    employees_df = pd.DataFrame(employees)
    employees_df_count_by_occ = employees_df.groupby(by=["occ_code"]).count()["name"]
    employees_rates_category_by_occupation = rates_category_by_occupation.loc[
        employees_df.groupby(by=["occ_code"]).count()["name"].index
    ]
    return employees_rates_category_by_occupation.multiply(
        employees_df_count_by_occ.values, axis=0
    )
```
