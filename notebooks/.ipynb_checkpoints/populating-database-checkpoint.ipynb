```python
def expected_number_of_injuries_per_agegroup_for_occupation(
    companyId=1
):
    """Returns the expected number of injuries per age group"""
    cnx = mysql.connector.connect(
        user="doadmin",
        password="00Cmib44qde23Wkj",
        host="db-mysql-sfo2-95612-do-user-9925891-0.b.db.ondigitalocean.com",
        port=25060,
        database="PrevWorks",
    )
    cursor = cnx.cursor()
    cursor.execute(
        f"""
        WITH t3 AS (
            SELECT
                t1.userId,
                t2.age,
                t2.occupationId
            FROM PrevWorks.user2company AS t1
            LEFT JOIN PrevWorks.user AS t2
            ON t1.userId = t2.userId AND t1.companyId = {companyId}
        ),
        t5 AS (
            SELECT
                t4.occupationId,
                t3.age,
                SUM(t4.probability) AS expected_number_of_injuries
            FROM t3
            LEFT JOIN PrevWorks.occupationAgeProbabilities AS t4
            ON t3.occupationId = t4.occupationId
            AND t3.age >= t4.ageLower
            AND t3.age < t4.ageUpper
            GROUP BY t3.occupationId, t3.age
            ORDER BY ASC
        )
        SELECT
            t6.occupationName,
            t5.avgDays,
            t5.expected_number_of_injuries
        FROM PrevWorks.occupations AS t6
        INNER JOIN t5
        ON t5.occupationId = t6.occupationId
        """
    )
    expected_number_of_injuries_for_company = cursor.fetchall()[0]
    cursor.close()
    cnx.close()
    return expected_number_of_injuries_for_company
```

```python
def expected_number_of_injuries_per_bodypart_for_occupation(
    companyId=1
):
    """Returns the expected number of injuries per body part in company"""
    cnx = mysql.connector.connect(
        user="doadmin",
        password="00Cmib44qde23Wkj",
        host="db-mysql-sfo2-95612-do-user-9925891-0.b.db.ondigitalocean.com",
        port=25060,
        database="PrevWorks",
    )
    cursor = cnx.cursor()
    cursor.execute(
        f"""
        WITH t3 AS (
            SELECT
                t1.userId,
                t2.occupationId
            FROM PrevWorks.user2company AS t1
            LEFT JOIN PrevWorks.user AS t2
            ON t1.userId = t2.userId AND t1.companyId = {companyId}
        ),
        t5 AS (
            SELECT
                t4.occupationId,
                t4.bodyId,
                SUM(t4.probability) AS expected_number_of_injuries
            FROM t3
            LEFT JOIN PrevWorks.occupationBodyProbability AS t4
            ON t3.occupationId = t4.occupationId
            GROUP BY t3.occupationId, t4.bodyId
            ORDER BY DESC
        ),
        t7 AS (
            SELECT
                t6.occupationName,
                t5.bodyId,
                t5.expected_number_of_injuries
            FROM PrevWorks.occupations AS t6
            INNER JOIN t5
            ON t5.occupationId = t6.occupationId
        )
        SELECT
            t7.occupationName,
            t8.name,
            t7.expected_number_of_injuries
        FROM PrevWorks.bodyParts AS t8
        INNER JOIN t7
        ON t7.bodyId = t8.bodyId
        """
    )
    expected_number_of_injuries_for_company = cursor.fetchall()[0]
    cursor.close()
    cnx.close()
    return expected_number_of_injuries_for_company
```

```python
def expected_number_of_injuries_per_avgdays_for_occupation(
    companyId=1
):
    """Returns the expected number of injuries """
    cnx = mysql.connector.connect(
        user="doadmin",
        password="00Cmib44qde23Wkj",
        host="db-mysql-sfo2-95612-do-user-9925891-0.b.db.ondigitalocean.com",
        port=25060,
        database="PrevWorks",
    )
    cursor = cnx.cursor()
    cursor.execute(
        f"""
        WITH t3 AS (
            SELECT
                t1.userId,
                t2.occupationId
            FROM PrevWorks.user2company AS t1
            LEFT JOIN PrevWorks.user AS t2
            ON t1.userId = t2.userId AND t1.companyId = {companyId}
        ),
        t5 AS (
            SELECT
                t4.occupationId,
                t4.avgDays,
                SUM(t4.probability) AS expected_number_of_injuries
            FROM t3
            LEFT JOIN PrevWorks.occupationDaysProbabilities AS t4
            ON t3.occupationId = t4.occupationId
            GROUP BY t3.occupationId, t4.avgDays
            ORDER BY DESC
        )
        SELECT
            t6.occupationName,
            t5.avgDays,
            t5.expected_number_of_injuries
        FROM PrevWorks.occupations AS t6
        INNER JOIN t5
        ON t5.occupationId = t6.occupationId
        """
    )
    expected_number_of_injuries_for_company = cursor.fetchall()[0]
    cursor.close()
    cnx.close()
    return expected_number_of_injuries_for_company
```
