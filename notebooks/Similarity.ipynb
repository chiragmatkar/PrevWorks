```python
from sentence_transformers import SentenceTransformer
import pandas as pd
```

```python
model = SentenceTransformer('sentence-transformers/paraphrase-MiniLM-L6-v2')
```

```python
kaggle_insurance_data = pd.read_csv("../data/Kaggle Actuarial loss prediction/train.csv")
data = pd.read_csv("../data/processed.csv", index_col=[0])
```

```python
kaggle_insurance_data
```

```python
for i in list(kaggle_insurance_data['ClaimDescription'].unique()):
    print(i)
```

```python
kaggle_desc = kaggle_insurance_data['ClaimDescription'].unique()
embeddings = model.encode(kaggle_desc)
print(embeddings)
```

```python
data.head()
```

```python
full_description = data['part_of_body'] + ' ' + data['event_type'] + ' ' + data['src_of_injury']
full_description = full_description.unique()
```

```python
embeddings_data = model.encode(full_description)
print(embeddings)
```

```python
len(full_description)
```

```python
len(embeddings_data)
```

```python
import numpy as np
from scipy.spatial import distance
from tqdm import tqdm

mapping_dict = dict()

for i in tqdm(range(len(embeddings))):
    distances = distance.cdist([embeddings[i]], embeddings_data, "cosine")[0]
    min_index = np.argmin(distances)
    mapping_dict[kaggle_desc[i]] = full_description[min_index]
```
